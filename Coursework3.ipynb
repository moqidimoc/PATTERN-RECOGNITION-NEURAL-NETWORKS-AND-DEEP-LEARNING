```python
# Here we'll apply the Delta Learning Algorithm to learn the weight of an ANN

# Let's start by importing the requested libraries for this problem
import pandas as pd
import numpy as np
```

```python
# First, let's see how it works on the example of the tutorial

# Let's make a simple program to apply the Delta Learning Algorithm
def seqDeltaL(data, epochs, w, eta): #eta is the learning rate
    for e in range(epochs): # for the number of epochs we specify
        print("Epoch #{}".format(e+1))
        for i in range(len(data)): # for all points in our dataset
            print("Iteration #{}".format(e*len(data)+i+1))

            # y=H(Wx)
            y =  np.dot(data.iloc[i,:(len(data.columns)-1)], w) # W*x
            if y < 0: y=0 
            elif y > 0: y=1
            elif y == 0: y=0.5 # This means that for the Heaviside function we have H(0)=0.5  if it's zero
            print("y=H(wx)={}".format(y))

            # Now we update w if there is a difference between t (desired ouput==data.iloc[i, len(data.columns)-1]) and y
            w = w + eta*(data.iloc[i, len(data.columns)-1]-y)*data.iloc[i, :(len(data.columns)-1)]
            print("w_new={} \n".format(w.values))

        print("End of epoch #{} \n".format(e+1))
        
    return w.values        
```

```python
# Tutorial problem (NOT)
data = pd.DataFrame([[1,0,1], [1,1,0]]) # datapoints are 0 and 1 with targets 1 and 0 respectively
w = np.array([-1.5, 2]) # threshold is 1.5 and w1 is 2
eta = 1
epochs = 5

new_w = seqDeltaL(data, epochs, w, eta)
print("Final weights: {}".format(new_w)) # Should be [ 0.5 -1. ]
```

```python
# You did great with a small dataset, now you should be able to perform this sequential Delta Learning Algorithm on big datasets
# too. The coursework asks you to perform it on the iris dataset, so let's start by downloading it
from sklearn import datasets
iris = datasets.load_iris()
print(iris.DESCR)
```

```python
# As our function performs on dataframes and numpy arrays, we will adapt this info to do so
irisdf=pd.DataFrame(iris.data)

irisdf.head(5)
```

```python
# We also have to use Augmented notation, so we have to add a column of 1s at the beginning of the dataframe (1st column=0)
ones = pd.Series([1]*len(irisdf)) # we create the vector
irisdf = pd.concat([ones, irisdf], axis=1) # we add it to the dataframe

irisdf.head(5) # Let's have a peak of how it looks
```

```python
# The last part is a little tricky, the coursework asks us to do a classifier of two classes (1 for class 0==setosa and 0 for 
# the other classes 1 & 2==versicolor and virginica)
y = pd.Series([1 if x==0 else 0 for x in iris.target]) # Here we create the target array
irisdf = pd.concat([irisdf, y], axis=1) # And here we add it to the last columns
irisdf.columns = [0,1,2,3,4,5] # we also rename the columns so there is some logic in them

irisdf.head(5)
```

```python
# Looks like we already have our data prepared, so let's perform the Delta Learning Algorithm on it:
data = irisdf
w = np.array([0.5,3.5,-1.5,-1.5,-3.5]) # threshold is -0.5, w1 is 3.5, w2 is -1.5, w3 is -1.5 andw4 is-3.5
eta = 0.1
epochs = 2

new_w = seqDeltaL(data, epochs, w, eta)
print("Final weights: {}".format(new_w)) # Should be [ 0.3   2.34 -1.83 -2.5  -3.8 ] 
```

```python
# Now that we have our new weighted vector, we can estimate how good did our classifier do, let's do this by evaluating it

# First we initialise the variables where we will put our scores
score_after = 0
for i in range(150): # for all points
    
    # y=H(Wx)
    y =  np.dot(data.iloc[i,:(len(data.columns)-1)], new_w) # W*x
    if y < 0: y=0 
    elif y > 0: y=1
    elif y == 0: y=0.5 # This means that for the Heaviside function we have H(0)=0.5  if it's zero
    
    if data.iloc[i, len(data.columns)-1] == y: score_after+=1 # The classifier predicted well
        
# Now that we have the instances that were correctly predicted, let's see the score in percentage:
total_score_after = score_after*100/len(data)

print(total_score_after)
```
